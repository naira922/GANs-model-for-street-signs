```python
import os
import random
import shutil

# Your original folder containing all images
image_folder = '/content/images'

# Creating destination folders for domain A and domain B
trainA = '/content/datasets/road_signs/trainA'
trainB = '/content/datasets/road_signs/trainB'

# Create the directories if they don't exist
os.makedirs(trainA, exist_ok=True)
os.makedirs(trainB, exist_ok=True)

# List all images in the original folder (only files, no directories)
image_list = [f for f in os.listdir(image_folder) if os.path.isfile(os.path.join(image_folder, f))]

# Shuffle the images to ensure random split
random.shuffle(image_list)

# Calculate split point for equal division
split_point = len(image_list) // 2

# Split the images into two groups
images_A = image_list[:split_point]  # First half for trainA
images_B = image_list[split_point:]  # Second half for trainB

# Copy images to the respective folders
for img in images_A:
    shutil.copy(os.path.join(image_folder, img), os.path.join(trainA, img))

for img in images_B:
    shutil.copy(os.path.join(image_folder, img), os.path.join(trainB, img))

print(f"Dataset successfully split into {len(images_A)} images for trainA and {len(images_B)} images for trainB.")
```

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Path to the generated images
generated_images_path = './results/road_signs_cyclegan/test_latest/images/'

# List all the images in the folder
generated_images = os.listdir(generated_images_path)

# Filter to show only images (if there are non-image files)
image_files = [img for img in generated_images if img.endswith('.png') or img.endswith('.jpg')]

# Show each image using matplotlib
for i, img_file in enumerate(image_files):
    img_path = os.path.join(generated_images_path, img_file)

    # Open the image
    img = Image.open(img_path)

    # Display the image
    plt.figure()
    plt.imshow(img)
    plt.axis('off')  # Hide axis
    plt.title(f'Generated Image {i+1}: {img_file}')
    plt.show()
```
